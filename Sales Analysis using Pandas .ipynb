## Exploratory Data Analysis

**What is overall trend?**

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#Getting month year from the dataset
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['month_year']
```

```python
#Grouping month year
df_trend=df.drop('year',axis=1, inplace=True)
df_trend=df.groupby('month_year').sum()
df_trend
```

```python
df_trend = df_trend.groupby('month_year').sum()['sales'].reset_index()
df_trend
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'],df_trend['sales'],color='#1db859')
plt.xticks(rotation=90)
plt.show()
```

 - **Which are the top 10 product by sales?**

```python
prod_sales=pd.DataFrame(df.groupby('product_name')['sales'].sum())
prod_sales=prod_sales.sort_values('sales',ascending=False)
```

```python
prod_sales[:10]
```

- **Which are the most selling products?**

```python
hs_prod=pd.DataFrame(df.groupby('product_name')['quantity'].sum())
hs_prod=hs_prod.sort_values('quantity',ascending=False)
hs_prod
```

```python
hs_prod[:50]
```

```python
hs_prod.tail(100)
```

 - **What is most prefered shiping mode?**

```python
plt.figure(figsize=(10,6))
sns.countplot(df['ship_mode'])
```

- **Which are most profitable category and sub category?**

```python
profitable_subcat=pd.DataFrame(df.groupby('sub_category')['profit'].sum())
profitable_subcat=profitable_subcat.sort_values('profit',ascending=False)
profitable_subcat
```

```python
df['sub_category'].unique()
```

```python
profitable_subcat_order=pd.DataFrame(df.groupby(['category','sub_category'])['profit'].sum())
profitable_subcat_order.sort_values(['category','profit'],ascending=False)
```
