```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python


df = pd.read_csv('District_health_data_in_india.csv')
df.head()
```

```python
df.columns
```

```python
state_vice_gender_ratio=df.groupby('State/UT')[' Sex ratio of the total population (females per 1,000 males)'].mean().reset_index().sort_values(' Sex ratio of the total population (females per 1,000 males)',ascending = False)['State/UT']
state_vice_gender_ratio
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
sns.barplot(x='State/UT',y=' Sex ratio of the total population (females per 1,000 males)',data=df,order=state_vice_gender_ratio,ax=ax)
plt.xticks(rotation=90)
```

### Above plot shows southern states have good amount of sex ratio(females count for 1000 males)

```python
df['Sex ratio at birth for children born in the last five years (females per 1,000 males)']=df['Sex ratio at birth for children born in the last five years (females per 1,000 males)'].astype(str).astype(float)
df.loc[323,'Sex ratio at birth for children born in the last five years (females per 1,000 males)']=1261
df.loc[330,'Sex ratio at birth for children born in the last five years (females per 1,000 males)']=1111
```

```python
df['Sex ratio at birth for children born in the last five years (females per 1,000 males)'].dtypes
```

```python
state_vice_child_gender_ratio=df.groupby('State/UT')['Sex ratio at birth for children born in the last five years (females per 1,000 males)'].mean().reset_index().sort_values('Sex ratio at birth for children born in the last five years (females per 1,000 males)',ascending = False)['State/UT']
state_vice_child_gender_ratio
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
sns.barplot(x='State/UT',y='Sex ratio at birth for children born in the last five years (females per 1,000 males)',data=df,order=state_vice_child_gender_ratio,ax=ax)
plt.xticks(rotation=90)
```

### Above plot shows southern states have poor for child sex ratio(females child count for 1000 males child)
